```python
import pandas as pd
df=pd.read_csv('D:\Data_file\Pokemon.csv')
```

### Making changes to the data

```python
df.head(3)
```

```python
df['Special']=df['Sp. Atk']+ df['Sp. Def']
print(df.head(3))
```

```python
#Dropping a column
df=df.drop(columns=['Special','Total'])
df.head(2)
```

```python
##another way to add a column is
df['total']=df.iloc[:,4:10].sum(axis=1) #axis=1 means horizontal,axis=0 means vertical
df.head(2)                              #: all the rows and 4:10 for columns [:,4:10]
```

```python
##Rearranging the columns
cols=list(df.columns)
df=df[cols[0:4]+[cols[-1]]+cols[4:12]]
df.head(3)
```

### Saving our data(Exporting in desired format)

```python
df.to_csv('data_modified.csv',index=False)
#for converting it to excel file
#df.to_excel('data_modified.xlsk',index=False)
#df.to_csv('modified.txt',index=False,sep='\t')    for text file
```

### Filtering Data

```python
df.loc[(df['Type 1'] == 'Grass') & (df['Type 2']== 'Poison')]  #in pandas df we use & instead of and
#here and:&  or:|  not:~
```

```python
new_df=df.loc[(df['Type 1'] == 'Grass') & (df['Type 2']== 'Poison') & (df['HP'] > 70)]
new_df
```

```python
new_df=new_df.reset_index(drop=True) #used for resetting the index
new_df
```

```python
not_mega=df.loc[~df['Name'].str.contains('Mega')]
not_mega   #Removing mega from the term
```

```python
import re     #here we are importing regex & it's a very powerful tool for string operations.
df.loc[df['Type 1'].str.contains('fire|grass',flags=re.I,regex=True)]  #re.I means ignoring the case
```

```python
df.loc[df['Name'].str.contains('pi[a-z]*',flags=re.I,regex=True)]  #* means zero or more
```

```python
df.loc[df['Name'].str.contains('^pi[a-z]*',flags=re.I,regex=True)]
```

```python
#Changing the fire type pokemon to flame 
df.loc[df['Type 1']== 'Fire','Type 1']='flamer'
df.head(10)
```

```python
#Changing it again to fire
df.loc[df['Type 1']=='flamer','Type 1']='Fire'
df.head(10)
```
